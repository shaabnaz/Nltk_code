```python
import cv2
import os

def images_to_video(image_folder, output_video, fps=5, repeat_frames=5):
    images = [img for img in os.listdir(image_folder) if img.endswith(('.png', '.jpg', '.jpeg'))]
    images.sort()  # Ensure the correct order
    
    if not images:
        print("No images found in the folder.")
        return
    
    first_image = cv2.imread(os.path.join(image_folder, images[0]))
    height, width, _ = first_image.shape
    
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4 output
    video_writer = cv2.VideoWriter(output_video, fourcc, fps, (width, height))
    
    for image in images:
        img_path = os.path.join(image_folder, image)
        frame = cv2.imread(img_path)
        if frame is None:
            print(f"Skipping invalid image: {image}")
            continue
        
        # Write the same frame multiple times to slow down playback
        for _ in range(repeat_frames):
            video_writer.write(frame)

    video_writer.release()
    print(f"Video saved as {output_video}")

# Example usage
image_folder = r"path)"
output_video = 'output_video.mp4'  # Output video file name
fps = 5  # Reduce FPS
repeat_frames = 15 # Repeat each frame multiple times to slow it down

images_to_video(image_folder, output_video, fps, repeat_frames)
```

```python
import cv2
import os
import numpy as np

def images_to_video_with_motion(image_folder, output_video, fps=10, step=5):
    """
    Converts images to a video with a walking motion effect.
    :param image_folder: Folder containing images.
    :param output_video: Output video file name.
    :param fps: Frames per second.
    :param step: Pixels to shift each frame (higher = faster movement).
    """
    images = sorted([img for img in os.listdir(image_folder) if img.endswith(('.png', '.jpg', '.jpeg'))])
    
    if not images:
        print("No images found in the folder.")
        return

    first_image = cv2.imread(os.path.join(image_folder, images[0]))
    height, width, _ = first_image.shape

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4 output
    video_writer = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

    for image in images:
        img_path = os.path.join(image_folder, image)
        frame = cv2.imread(img_path)

        if frame is None:
            print(f"Skipping invalid image: {image}")
            continue

        # Simulating walking by shifting the image
        for shift in range(0, width, step):  # Move image from left to right
            moved_frame = np.roll(frame, shift, axis=1)  # Shift horizontally
            video_writer.write(moved_frame)

    video_writer.release()
    print(f"🎬 Video saved as {output_video}")

# Example usage
image_folder = r"path"
#image_folder = "path_to_images_folder"  # Replace with your folder path
output_video = "walking_effect_video.mp4"  # Output video file name
fps = 10  # Frames per second
step = 10  # Pixels per frame shift

images_to_video_with_motion(image_folder, output_video, fps, step)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Provide correct path to your image
path = r"path"

# Load your CXR image in grayscale
image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

# Check if image loaded properly
if image is None:
    print("Image not found at given path.")
else:
    # Generate a fake binary mask using thresholding
    _, binary_mask = cv2.threshold(image, 100, 1, cv2.THRESH_BINARY)

    # Apply the mask
    masked_image = image * binary_mask

    # Visualize
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 3, 1)
    plt.title("Original Image")
    plt.imshow(image, cmap="gray")

    plt.subplot(1, 3, 2)
    plt.title("Fake Mask")
    plt.imshow(binary_mask, cmap="gray")

    plt.subplot(1, 3, 3)
    plt.title("Masked Image")
    plt.imshow(masked_image, cmap="gray")

    plt.tight_layout()
    plt.show()
```
